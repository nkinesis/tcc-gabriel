```python
import pandas as pd
import numpy as np
```

```python
url = r'C:\Users\ullma\Documents\Google_Drive\Desktop\Faculdade\2020-1\PTCC\data\olist-processed\new2.csv'
original_items = pd.read_csv(url, sep=',')
items_filtered_cols = original_items[['encoded_id', 'order_approved_at_month', 'product_id']]
items_filtered_cols
```

```python
unique_products = np.unique(items_filtered_cols[['product_id']].values)
len(unique_products)
```

```python
items_filtered_cols['product_id'] = items_filtered_cols['product_id'].astype('category')
```

```python
items_filtered_cols['product_id'].dtypes
```

```python
items_filtered_cols['product_id'] = items_filtered_cols['product_id'].cat.codes
count = pd.value_counts(items_filtered_cols['product_id'].values)
count_matrix = count.reset_index().values.tolist()
count_matrix[:100]
```

```python
items_filtered_cols
```

```python
products_to_fetch = np.array([])
limit_x_products = 10
current_product = 0
for row in count_matrix:
    products_to_fetch = np.append(products_to_fetch, int(row[0]))
    current_product += 1
    if (current_product >= limit_x_products):
        break
products_to_fetch
```

```python
fetched_products = items_filtered_cols[(items_filtered_cols.product_id.isin(products_to_fetch))]
fetched_products.shape
```

```python
items_filtered_cols = fetched_products
dum_df = pd.get_dummies(items_filtered_cols, columns=["product_id"], prefix=["product"] )
dum_df
```
