```python
import pandas as pd
import numpy as np
from numpy import array
from numpy import argmax
from keras.models import load_model
```

```python
def greatestValue(array):
    greatest = -1
    counter = 0
    positionFound = 0
    for value in array:
        if value > greatest:
            greatest = value
            positionFound = counter
        counter += 1
        
    return [greatest, positionFound]
```

```python
# load real outputs and encode
url = './test-uci.csv'
real_ds = pd.read_csv(url, delimiter=',')
real_outputs = pd.get_dummies(real_ds.modeProd, prefix='prod')
real_outputs[:3]
```

```python
# load real inputs and encode
real_inputs = pd.get_dummies(real_ds.clientId, prefix='client')
num_inputs = len(real_inputs.columns) + 1
real_inputs['month'] = real_ds.month
real_inputs = real_inputs.astype('uint8')
real_inputs[:3]
```

```python
predictions = []
for i in range(len(real_inputs)):
    #select a row to use in the prediction and predict
    row = real_inputs.values[i]
    row = row.reshape((1, len(row)))
    loaded_model = load_model('./model2020-08-08_12-09-12.h5')
    outputs = loaded_model.predict(row, verbose=0)
    #print(np.shape(outputs[0]))

    # discover which product was predicted
    prod_names = list(real_outputs)
    prediction_max = np.argmax(outputs[0])
    prod_predicted = prod_names[prediction_max]
    predictions.append(prod_predicted)
    
predictions
```

```python
real = real_ds.modeProd.values
real[:3]
```

```python
correct_predictions = 0 
incorrect_predictions = 0 
for row in range(len(predictions)):
    if real[row] == 'prod_' + str(predictions[row]):
        correct_predictions += 1
    else:
        incorrect_predictions += 1
print(correct_predictions)
print(incorrect_predictions)
```
